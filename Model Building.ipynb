```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/mobile_phone_dataset_cleaned.csv')
```

```python
df
```

```python
df1 = df.drop(columns = ['Unnamed: 0.1','Unnamed: 0','Name','No_of_sim','Camera','External_Memory','Screen_resolution'])
```

```python
df1
```

```python
from sklearn.preprocessing import MinMaxScaler,StandardScaler
```

```python
min_max = MinMaxScaler()
```

```python
standard = StandardScaler()
```

```python
df1_scaled = df1.copy()
df1_scaled
```

```python
cols = ['Battery','Price','Inbuilt_memory','hresolution','vresolution','External_memory']
```

```python
df1_scaled[cols] = standard.fit_transform(df1[cols])
```

```python
df1_scaled
```

```python
from sklearn.model_selection import train_test_split
```

```python
x = df1_scaled.drop(columns = ['Spec_score'],axis = 1)
y = df1_scaled['Spec_score']
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
y_pred = lr.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error,r2_score, mean_absolute_error
```

```python
r2_score(y_test,y_pred)
```

```python
mean_absolute_error(y_test,y_pred)
```

```python
mean_squared_error(y_test,y_pred)
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
dt = DecisionTreeRegressor()
```

```python
dt.fit(x_train,y_train)
```

```python
y_pred = dt.predict(x_test)
```

```python
r2_score(y_test,y_pred)
```

```python
mean_absolute_error(y_test,y_pred)
```

```python
mean_squared_error(y_test,y_pred)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rf = RandomForestRegressor()
```

```python
rf.fit(x_train,y_train)
```

```python
y_pred = rf.predict(x_test)
```

```python
r2_score(y_test,y_pred)
```

```python
mean_absolute_error(y_test,y_pred)
```

```python
mean_squared_error(y_test,y_pred)
```

Hypreparameter Tuning

Decision Tree

```python
max_depth = [3,4,5]
min_samples_split = [4,6,8]
min_samples_leaf = [4,5,6]
```

```python
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV
```

```python
dt1 = DecisionTreeRegressor()
```

```python
param_grid = {'max_depth':max_depth,'min_samples_split':min_samples_split, 'min_samples_leaf':min_samples_leaf}
```

```python
grid_search = GridSearchCV(estimator=dt1, param_grid=param_grid, cv=5, n_jobs=-1, verbose=1)
```

```python
grid_search.fit(x_train,y_train)
```

```python
best_model = grid_search.best_estimator_
y_pred = best_model.predict(x_test)
```

```python
r2_score(y_test,y_pred)
```
